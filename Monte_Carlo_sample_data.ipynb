```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("simulated_mc.csv")
df.head()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["RollingReturn"], label="Rolling Avg Return", color='orange', linewidth=2)
plt.title("Rolling Average Return (Every Visit )")
plt.xlabel("Episode")
plt.ylabel("Normalized Return")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["RollingSuccess"], label="Rolling Success Rate", color='green', linewidth=2)
plt.title("Rolling Success Rate (Reward > 0.6)")
plt.xlabel("Episode")
plt.ylabel("Success Rate")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# Plot Rolling Success Rate
plt.figure(figsize=(10, 5))
plt.plot(df["RollingSuccess"].dropna(), label="Success Rate", linewidth=2, color='darkorange')
plt.title("Rolling Success Rate ")
plt.xlabel("Episode")
plt.ylabel("Success Rate")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
df.groupby("Event")["Reward"].plot(kind='kde', legend=True)
plt.title("Reward Distribution ")
plt.xlabel("Reward")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 5))
df["Event"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Frequency")
plt.ylabel("Count")
plt.xlabel("Event Type")
plt.grid(True)
plt.tight_layout()
plt.show()
```

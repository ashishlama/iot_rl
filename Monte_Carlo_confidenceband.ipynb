```python
import numpy as np
import gymnasium as gym
import random
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from collections import defaultdict
import time
import os
```

```python
def epsilon_greedy(Q, state, nA, epsilon):
    if random.random() < epsilon:
        return random.choice(np.arange(nA))
    return np.argmax(Q[state])
```

```python
def mc_control_first_visit(env, episodes=5000, gamma=0.99):
    returns_sum, returns_count = defaultdict(float), defaultdict(float)
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    rewards, success = [], []
    start = time.time()

    for i in range(episodes):
        epsilon = max(0.05, 1.0 - i / episodes)
        state, _ = env.reset()
        episode, done = [], False

        while not done:
            action = epsilon_greedy(Q, state, env.action_space.n, epsilon)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            episode.append((state, action, reward))
            state = next_state

        G, visited = 0, set()
        for t in reversed(range(len(episode))):
            s, a, r = episode[t]
            G = gamma * G + r
            if (s, a) not in visited:
                visited.add((s, a))
                returns_sum[(s, a)] += G
                returns_count[(s, a)] += 1
                Q[s][a] = returns_sum[(s, a)] / returns_count[(s, a)]

        rewards.append(sum(x[2] for x in episode))
        success.append(1 if episode[-1][2] == 1.0 else 0)

    return Q, rewards, success, time.time() - start
```

```python
def mc_control_every_visit(env, episodes=5000, gamma=0.99):
    returns_sum, returns_count = defaultdict(float), defaultdict(float)
    Q = defaultdict(lambda: np.zeros(env.action_space.n))
    rewards, success = [], []
    start = time.time()

    for i in range(episodes):
        epsilon = max(0.05, 1.0 - i / episodes)
        state, _ = env.reset()
        episode, done = [], False

        while not done:
            action = epsilon_greedy(Q, state, env.action_space.n, epsilon)
            next_state, reward, terminated, truncated, _ = env.step(action)
            done = terminated or truncated
            episode.append((state, action, reward))
            state = next_state

        G = 0
        for t in reversed(range(len(episode))):
            s, a, r = episode[t]
            G = gamma * G + r
            returns_sum[(s, a)] += G
            returns_count[(s, a)] += 1
            Q[s][a] = returns_sum[(s, a)] / returns_count[(s, a)]

        rewards.append(sum(x[2] for x in episode))
        success.append(1 if episode[-1][2] == 1.0 else 0)

    return Q, rewards, success, time.time() - start
```

```python
# Run agents
env = gym.make("FrozenLake-v1", map_name="8x8", is_slippery=True)
Qf, rf, sf, tf = mc_control_first_visit(env)
Qe, re, se, te = mc_control_every_visit(env)
```

```python
# Plots
def plot_metric(values, label, ylabel):
    rolling = pd.Series(values).rolling(100).mean()
    plt.plot(rolling, label=label)
    plt.xlabel("Episode")
    plt.ylabel(ylabel)
    plt.grid(True)
    plt.legend()
    plt.show()

plot_metric(rf, "First Visit - Return", "Return")
plot_metric(sf, "First Visit - Success Rate", "Success Rate")
plot_metric(re, "Every Visit - Return", "Return")
plot_metric(se, "Every Visit - Success Rate", "Success Rate")
```

```python
np.random.seed(42)
episodes = 500
decay_epsilon = np.linspace(1.0, 0.05, episodes)
```

```python
returns_first = np.clip(np.cumsum(np.random.normal(0.001, 0.02, episodes)), 0, 1)
returns_every = np.clip(np.cumsum(np.random.normal(0.0012, 0.02, episodes)), 0, 1)
```

```python
success_first = (returns_first > 0.6).astype(int)
success_every = (returns_every > 0.6).astype(int)
```

```python
df = pd.DataFrame({
    "Episode": np.arange(episodes),
    "Epsilon": decay_epsilon,
    "Return_FirstVisit": returns_first,
    "Return_EveryVisit": returns_every,
    "Success_FirstVisit": success_first,
    "Success_EveryVisit": success_every,
    "RollingReturn_FirstVisit": pd.Series(returns_first).rolling(50).mean(),
    "RollingReturn_EveryVisit": pd.Series(returns_every).rolling(50).mean(),
    "RollingSuccess_FirstVisit": pd.Series(success_first).rolling(50).mean(),
    "RollingSuccess_EveryVisit": pd.Series(success_every).rolling(50).mean()
})
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["RollingReturn_FirstVisit"], label="First Visit MC", linewidth=2)
plt.plot(df["RollingReturn_EveryVisit"], label="Every Visit MC", linewidth=2)
plt.title("Rolling Average Return (500 Episodes)")
plt.xlabel("Episode")
plt.ylabel("Average Return")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df["RollingSuccess_FirstVisit"], label="First Visit MC", linewidth=2)
plt.plot(df["RollingSuccess_EveryVisit"], label="Every Visit MC", linewidth=2)
plt.title("Rolling Success Rate (500 Episodes)")
plt.xlabel("Episode")
plt.ylabel("Success Rate")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df["Return_FirstVisit"][-100:], bins=15, alpha=0.6, label="First Visit MC")
plt.hist(df["Return_EveryVisit"][-100:], bins=15, alpha=0.6, label="Every Visit MC")
plt.title("Return Distribution (Last 100 Episodes)")
plt.xlabel("Return")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
for agent in ["FirstVisit", "EveryVisit"]:
    rolling = df[f"Return_{agent}"].rolling(50)
    mean = rolling.mean()
    std = rolling.std()

    plt.figure(figsize=(10, 5))
    plt.plot(mean, label=f"{agent} - Mean Return", color='black' if agent == "FirstVisit" else 'black')
    plt.fill_between(df["Episode"], mean - std, mean + std, alpha=0.2,
                     color='red' if agent == "FirstVisit" else 'yellow')
    plt.title(f"{agent} - Return with Confidence Band")
    plt.xlabel("Episode")
    plt.ylabel("Return")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
```
